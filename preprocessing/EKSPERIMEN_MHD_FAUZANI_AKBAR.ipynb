### **Hapus Missing Value**

```python
test = test.dropna()
test.info()
```

### **Hapus Duplicate Value**

```python
df = df.drop_duplicates()
test = df.drop_duplicates()
print(df.duplicated().sum())
print(test.duplicated().sum())
```

### **Standarisasi**

```python
scaler = StandardScaler()
df[num] = scaler.fit_transform(df[num])
test[num] = scaler.fit_transform(test[num])
df[num].head()
test[num].head()
```

### **Onehot Encoder**

```python
ohe = OneHotEncoder(sparse_output=False)
df_encoded = pd.DataFrame(ohe.fit_transform(df[cat]), columns=ohe.get_feature_names_out(cat))
test_encoded = pd.DataFrame(ohe.fit_transform(test[cat]), columns=ohe.get_feature_names_out(cat))
df = pd.concat([df, df_encoded], axis=1)
test = pd.concat([test, test_encoded], axis=1)
df = df.drop(columns=cat)
test = test.drop(columns=cat)
df.head()
test.head()
```
